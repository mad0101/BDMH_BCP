### Use CountVectorizer and Test Accuracy for the Test Set in the Train-Test Split is 72%

```python
# Define the training samples and labels
#samples = ['AAAFVNQHLCGSHLVEAL', 'AACFKEGSFDEDGFYAKV', 'AAEELMNALVNSTLLETRT', 'AAGTAAGGKNFVIQKSEKSE']
#labels = [0, 0, 0, 0]

# Split the samples and labels into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(samples, labels, test_size=0.2, random_state=42)

# Convert the training samples into feature vectors using the CountVectorizer
vectorizer = CountVectorizer(analyzer='char')
X_train_vec = vectorizer.fit_transform(X_train)

print(X_train_vec.shape)

# Train a Support Vector Machine (SVM) classifier on the training data
clf = SVC(kernel='linear', C=1, random_state=42)
clf.fit(X_train_vec, y_train)

# Convert the testing samples into feature vectors using the CountVectorizer
X_test_vec = vectorizer.transform(X_test)
#-----------------Conversion---------------------------#
X_final_test_vec = vectorizer.transform(test_samples)

print(X_test_vec.shape)

# Use the trained classifier to predict the labels of the testing data
y_pred = clf.predict(X_test_vec)

#----------------Conversion----------------------------#
y_pred_final = clf.predict(X_final_test_vec)

# Evaluate the accuracy of the classifier on the testing data
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train_table,labels, test_size=0.2, random_state=42)

# Define the pipeline steps
scaler = StandardScaler()
classifier = RandomForestClassifier()

pipeline = Pipeline([
    ('scaler', scaler),
    ('classifier', classifier)
])

# Train the pipeline on the training data
pipeline.fit(X_train, y_train)

# Evaluate the pipeline on the testing data
score = pipeline.score(X_test, y_test)
print(f"Accuracy: {score:.2f}")

# Use the trained classifier to predict the labels of the testing data
y_pred = pipeline.predict_proba(X_test)[:, 1]
y_pred = (y_pred > threshold).astype(int)
roc_score = roc_auc_score(y_test, y_pred)
print("ROC score: ", roc_score)
```

```python
import lightgbm as lgb
# Define model parameters
X_train, X_test, y_train, y_test = train_test_split(X_train_table,labels, test_size=0.2, random_state=42)

params = {
    'n_estimators': 1000,
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'max_bin':500,
    'num_leaves': 800,
    'learning_rate': 0.1,
    'feature_fraction': 0.9
}

# Create training and validation datasets for LightGBM
train_data = lgb.Dataset(X_train, label=y_train)
val_data = lgb.Dataset(X_test, label=y_test)

# Train model
model = lgb.train(params, train_data, num_boost_round=1000, valid_sets=[train_data, val_data], early_stopping_rounds=50)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate performance
threshold = 0.5
y_pred_binary = np.where(y_pred > threshold, 1, 0)
accuracy = (y_pred_binary == y_test).mean()
print('Accuracy:', accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, StackingClassifier
```

```python
# Define the base models
base_models = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('et', ExtraTreesClassifier(n_estimators=100, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=100, random_state=42))
]

# Define the stacking model
stacked_model = StackingClassifier(estimators=base_models, final_estimator=RandomForestClassifier(n_estimators=100, random_state=42))

# Train the pipeline on the training data
stacked_model.fit(X_train, y_train)

# Evaluate the pipeline on the testing data
score = stacked_model.score(X_test, y_test)
print(f"Accuracy: {score:.2f}")

# Use the trained classifier to predict the labels of the testing data
y_pred = stacked_model.predict_proba(X_test)[:, 1]
y_pred = (y_pred > threshold).astype(int)
roc_score = roc_auc_score(y_test, y_pred)
print("ROC score: ", roc_score)
```
